```python
import tkinter as tk
from tkinter import messagebox
import math

def evaluate_expression(expression):
    try:
        return eval(expression)
    except Exception as e:
        messagebox.showerror("Error", f"Error evaluating expression: {e}")
        return None

def click_button(value):
    current_text = str(entry.get())
    if value == "=":
        result = evaluate_expression(current_text)
        if result is not None:
            entry.delete(0, tk.END)
            entry.insert(tk.END, result)
    elif value == "C":
        entry.delete(0, tk.END)
    elif value == "sqrt":
        entry.insert(tk.END, "math.sqrt(" + current_text + ")")
    elif value == "^":
        entry.insert(tk.END, "**")
    elif value == "sin":
        entry.insert(tk.END, "math.sin(math.radians(" + current_text + "))")
    elif value == "cos":
        entry.insert(tk.END, "math.cos(math.radians(" + current_text + "))")
    elif value == "tan":
        entry.insert(tk.END, "math.tan(math.radians(" + current_text + "))")
    else:
        entry.insert(tk.END, value)
        
app = tk.Tk()
app.title("Calculator")
entry = tk.Entry(app, width=40, borderwidth=5)
entry.grid(row=0, column=0, columnspan=4)

buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+',
    '^', 'C', ' ',
    'sin', 'cos', 'tan'
]

row_val = 1
col_val = 0

for button in buttons:
    if button == ' ':
        continue
    if button == 'sqrt':
        b = tk.Button(app, text=button, width=5, height=2, command=lambda: click_button(button))
    elif button == '^':
        b = tk.Button(app, text=button, width=5, height=2, command=lambda: click_button(button))
    elif button == 'C':
        b = tk.Button(app, text=button, width=5, height=2, command=lambda: click_button(button))
    elif button in ['sin', 'cos', 'tan']:
        b = tk.Button(app, text=button, width=5, height=2, command=lambda b=button: click_button(b))
    else:
        b = tk.Button(app, text=button, width=5, height=2, command=lambda b=button: click_button(b))
    
    b.grid(row=row_val, column=col_val)
    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1

app.mainloop()
```
